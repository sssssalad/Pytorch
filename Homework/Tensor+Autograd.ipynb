##### (3)基于现有数据的创建

```python
x=torch.tensor([5.5,3])
print(x,x.dtype)
x=x.new_ones(4,3,dtype=torch.double)
#或#x=torch.ones(4,3,dtype=torch.double)
print(x)
x=torch.randn_like(x,dtype=torch.float)
#重置数据类型为服从N(0,1)正态分布的随机数
print(x)#产生的tensor的size相同
print(x.size(),x.shape)
```

### 3.2 张量的操作
##### （1）加法

```python
x=torch.randn(4,3)#N(0,1)正态分布
y=torch.rand(4,3)#[0,1)均匀分布
print(x+y)
print(torch.add(x,y))
y.add_(x)
print(y)
```

##### (2)索引操作（索引结果与元数据共享内存，会一同修改）

```python
x=torch.rand(4,3)
print(x[0,:])#第二列
y=x[0,:]
y+=1
print(y)
print(x[0,:])#源tensor也被+1
```

```python
x=torch.tensor([[1,1,1],[2,2,2],[3,3,3]])
print(x)
print(x[0,:])
y.copy_(x[0,:])#复制操作源数据不会改变
y+=1
print(y)
print(x[0,:])
```

### （3）维度变换
torch.view()返回的新tensor和源tensor共享内存，仅改变了这个张量的观察角度

torch.reshape()返回一个新的tensor，不改变源tensor

```python
x=torch.randn(4,4)
y=x.view(16)
z=x.view(-1,8)#-1:这一维度由其他维度决定
print(x.size(),y.size(),z.size())
x+=1
print(x)
print(y)#也加了1
```

```python
a=torch.randn(4,4)
b=x.reshape(16)
c=x.reshape(-1,8)#-1:这一维度由其他维度决定
print(a.size(),b.size(),c.size())
a+=1
print(a)
print(b)#没加1
```

### 3.3 广播机制

### Autograd使用
Autograd为张量的所有操作提供了自动求导机制

requires_grad: 如果需要为张量计算梯度，则为 True，否则为 False。 使用PyTorch 创建 tensor 时，可以指定 requires_grad 为 True(默认为 False)

grad_fn:用来记录变量是怎么来的，方便计算梯度。 例如 y=x*3，grad_fn 记录了 y 由 x 计算的过程。

grad:当执行完了 backward()函数之后，通过 x.grad 查看 x 的梯度值。

```python
#lg:z=(x+2)**2+3,求解函数z关于自变量x在x=2上的导数
x=torch.tensor([2.0],requires_grad=True)
y=x+2
z=y**2+3
z
```

```python
#使用自动求导
z.backward()
x.grad
```

#### 练习题

```python
import numpy as np
x=np.linspace(0,10,100)
x1=torch.tensor(x,requires_grad=True)
#torch.linspace (start, end, steps=100, out=None)
#返回一个1维张量，包含在区间start和end上均匀间隔的step个点
y1= torch.sin(x1)
y1.sum().backward()#对输入的tensor数据的某一维度求和
z = x1.grad
```

```python
plt.xlabel ( 'x')
plt.ylabel ( 'y')
plt.title ( "y=sin(x)")
plt.plot(x,np.sin(x)) 
plt.show ()
```

```python
plt.xlabel ( 'x')
plt.ylabel ( 'y')
plt.title ( "y=cos(x)")
plt.plot(x,z) 
plt.show ()
```
