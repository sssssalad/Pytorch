Tensor练习题3

```python
# 创建高斯随机噪声
mean = 0
sigma = 0.25
noise = sigma*torch.randn(img_tensor.size())+mean
#print(noise.size())
#print(noise)
```

```python
img_tensor_norm = img_tensor / 255 
noisy_img_tensor = (img_tensor_norm + noise)*255
noisy_img_tensor>255
# 将像素值变换到0到255之间
noisy_img_tensor *= (noisy_img_tensor>0)
noisy_img_tensor = noisy_img_tensor * (noisy_img_tensor<=255) + 255 * (noisy_img_tensor>255)

# 转化成整形数据
noisy_img_tensor = noisy_img_tensor.byte()
# print(noisy_img_tensor.size())
noisy_img_tensor.dtype
```

```python
plt.imshow(noisy_img_tensor)
plt.show()
```

Autograd练习题

```python
x = torch.linspace(0, 10, 100, requires_grad=True)
```

```python
x = torch.linspace(0, 10, 100, requires_grad=True)
y = torch.sin(x)
z = y.sum()
z.backward()
```

```python
x.requires_grad_(False)
plt.plot(x,torch.sin(x),label="sin(x)")
plt.plot(x,x.grad,label="gradient of sin(x)")
plt.legend()
plt.show()
```
